Dataframe

```python
df = pd.read_csv("bootcamp_large_dataset_with_issues.csv")
df.head()
```

```python
df.shape
df.info()
```

Missing values

```python
# Number of missing values by column
df.isnull().sum()
```

```python
# Filling the data automatically - With numerical data
df["Sales"].fillna(df["Sales"].mean(), inplace = True)
```

```python
# Filling the data automatically - With text columns
df["Category"].fillna("Unknown", inplace = True)
```

Duplicates

```python
df.duplicated().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

Changing the column names & type

```python
# Changing the type to make it more efficient
df["Category"] = df["Category"].astype("category")
```

Adding features

```python
df["Profit_Margin"] = df["Profit"] / df["Sales"]
df.info()
```

```python
df["Sales_per_unit"] = df["Sales"] / df["Quantity"]
# How much reveneu we get
```

```python
# To categorise values into business quantities
df["Sales_Category"] = df["Sales"].apply(lambda x: "High" if x > 3000 else "Medium" if x > 1000 else "Low")
df.head()
```

Filtering data

```python
df_filter = df[df["Profit"] > 0]
df_filter.head()
```

Aggregation

```python
# Check Sales column by Region column and I want the total values of sales
df.groupby("Region")["Sales"].sum()
```

```python
# Check Sales column by Category column and I want the total values of sales
df.groupby("Category")["Sales"].sum()
```

Pipeline - Export dataset

```python
# Saved in the Colab Folder or, if logged out, in the system
df.to_csv("cleaned_data.csv", index = False)
```
